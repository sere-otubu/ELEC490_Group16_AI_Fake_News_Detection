Script to get new NGROK url

```python
!pip install pyngrok
!curl -fsSL https://ollama.com/install.sh | sh

from google.colab import userdata
from pyngrok import ngrok
import os
import subprocess
import time

# 1. Retrieve the secret securely
try:
    NGROK_TOKEN = userdata.get('NGROK_TOKEN')
    print("✅ Ngrok token retrieved successfully.")
except userdata.SecretNotFoundError:
    print("❌ Error: NGROK_TOKEN not found in Colab Secrets!")

ngrok.kill()
ngrok.set_auth_token(NGROK_TOKEN)

# 2. Start Ollama Server
os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'
subprocess.Popen(["ollama", "serve"])
print("⏳ Starting Ollama server...")
time.sleep(10)

# 3. Pull Verified Models
print("📥 Pulling models (this may take 2-3 minutes)...")
!ollama pull gemma3:4b
!ollama pull embeddinggemma

# 4. Open Tunnel
public_url = ngrok.connect(11434, "http")
print(f"Update your backend config to: {public_url.public_url}")
```
