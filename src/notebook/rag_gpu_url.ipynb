Script to get new NGROK url

```python
import os
import subprocess
import time
from pyngrok import ngrok

# 1. Setup Auth
NGROK_TOKEN = "YOUR_TOKEN" 
ngrok.kill() # Clear any existing broken tunnels
ngrok.set_auth_token(NGROK_TOKEN)

# 2. Start Ollama Server
os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'
subprocess.Popen(["ollama", "serve"])
time.sleep(5)

# 3. Pull Verified Models (Check for typos here)
!ollama pull gemma3:4b
!ollama pull embeddinggemma # Replacing the suspected broken manifest

# 4. Open Tunnel
public_url = ngrok.connect(11434, "http")
print(f"Update your backend config to: {public_url.public_url}")
```
